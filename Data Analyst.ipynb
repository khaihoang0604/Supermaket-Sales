Câu 1: Bạn hãy xây dựng một hàm nhận tham số truyền vào là 1 chuỗi ký tự bất kỳ. Hàm kiểm tra và trả về kết quả
chuỗi trên là một đoạn password hợp lệ hay không, biết rằng password hợp lệ cần thoả mãn các tiêu chí:
- Chứa 1 ký tự đặc biệt trong các ký tự !, @, #, $,
- Chứa ít nhất một ký tự số
^
- Chứa ít nhất một ký tự chữ cái viết hoa
- Chứa ít nhất một ký tự chữ cái viết thường
- Độ dài từ 8 - 15 ký tự.

```python
def is_valid_password(password):
    # Khai báo các biến để kiểm tra các điều kiện
    has_special_char = False
    has_digit = False
    has_upper = False
    has_lower = False
    has_length = False
    # Duyệt qua từng ký tự trong mật khẩu
    special_chars = "!@#$"
    for char in password:
      # Kiểm tra ký tự đặc biệt
        if char in special_chars:
            has_special_char = True
      # Kiểm tra ký tự số
        if char.isdigit():
            has_digit = True
      # Kiểm tra ký tự viết hoa
        if char.isupper():
            has_upper = True
      # Kiểm tra ký tự viết thường
        if char.islower():
            has_lower = True

    # Kiểm tra độ dài chuỗi
    if 8<=len(password)<=15:
        has_length = True

    # Kiểm tra tất cả các điều kiện
    if has_special_char and has_digit and has_upper and has_lower and has_length:
        return print("Mật khẩu hợp lệ")
    else:
        return print("Mật khẩu không hợp lệ")

# Kiểm tra ví dụ
password = "Aa12345!"
is_valid_password(password)
```

Câu 3: Bạn hãy xây dựng một hàm nhận tham số truyền vào là một danh sách bất kỳ chỉ chứa các số. Hàm tính toán
và trả về số lớn nhì trong danh sách. Với trường hợp danh sách gồm các số bằng nhau, hàm trả về số lớn nhất. Nếu
có nhiều giá trị bằng số lớn nhì trong danh sách, hàm trả về 1 giá trị duy nhất. Ví dụ:
- Với danh sách [1,1,1,1,1,1] -> Hàm trả về 1
- Với danh sách [1,2,1,1,2,3] -> Hàm trả về 2

```python
def second_largest(nums):
    # Loại bỏ các số trùng lặp bằng cách chuyển danh sách thành set
    unique_nums = list(set(nums))

    # Kiểm tra nếu danh sách có ít hơn 2 số khác nhau
    if len(unique_nums) < 2:
        return unique_nums[0]

    # Sắp xếp danh sách giảm dần và lấy số lớn nhì
    unique_nums.sort(reverse=True)

    # Trả về số lớn nhì
    return unique_nums[1]

# Kiểm tra ví dụ
print(second_largest([1, 1, 1, 1, 1, 1]))
print(second_largest([1, 2, 1, 1, 2, 3]))
```

```python
import pandas as pd
```

```python
# B. Bạn hãy tìm ra các đơn hàng mua sản phẩm là ‘Health and beauty ’ hoặc ‘Sports and travel’
df[(df['Product line'] == 'Health and beauty') | (df['Product line'] == 'Sports and travel')]
```

```python
# C. Cột InvoiceID, giá trị được lưu trữ theo dạng XXX-StoreID-XXXX.
# Bạn hãy tạo 1 dataframe mới gồm các cột như trên file và tạo thêm một cột StoreID được tách từ InvoiceID.
# Sau đó tính tổng doanh số của từng StoreID đó.
df_Store = df
df_Store['StoreID'] = df['Invoice ID'].str.split('-').str[1]
df_Store
```
